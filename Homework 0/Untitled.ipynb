```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("D3.csv")

# Compute each of the explanitory values seperately
X1 = np.array(df.values[:,0])
X2 = np.array(df.values[:,1])
X3 = np.array(df.values[:,2])
y = np.array(df.values[:,3])

m = len(y) # Training samples
```

```python
def calculate_cost(X,y,theta):
    """Computes the cost function for linear regression"""
    predictions = X.dot(theta)
    errors = np.subtract(predictions,y)
    sqrErrors = np.square(errors)
    J= 1/(2*m)*np.sum(sqrErrors)
    return J

def gradient_descent(X,y,theta,alpha,iterations):
    """Computes the gradient descent for linear regression"""
    cost_history=np.zeros(iterations)

    for i in range(iterations):
        predictions = X.dot(theta)
        errors = np.subtract(predictions,y)
        sum_delta = (alpha/m)*X.transpose().dot(errors);
        theta = theta - sum_delta;
        cost_history[i] = calculate_cost(X,y,theta)
    
    return cost_history, theta
    
```

```python
# Problem 1
# Calculating linear regression model for X1, X2, X3 seperately
# Matrix with a single column of ones
X0 = np.ones((m,1))

# Reshaping X1, X2, and X3 to be a 2-D Array
X1=X1.reshape(m,1)
X2=X2.reshape(m,1)
X3=X3.reshape(m,1)
# Combining X0 with X1, X2, and X3 to stack them horizontally
X1=np.hstack((X0,X1))
X2=np.hstack((X0,X2))
X3=np.hstack((X0,X3))
```

```python
theta = np.zeros(2) # Theta_0 and theta_1

# Computing the inital cost for X1, X2, and X3
cost1 = calculate_cost(X1,y,theta)
cost2 = calculate_cost(X2,y,theta)
cost3 = calculate_cost(X3,y,theta)
```

```python
# Computing the gradient descent for X1, X2, and X3 seperately
alpha = .01
iterations = 1250

cost1_history, theta1 = gradient_descent(X1,y,theta,alpha,iterations)
iterations = 60
cost2_history, theta2 = gradient_descent(X2,y,theta,alpha,iterations)
iterations = 1250

cost3_history, theta3 = gradient_descent(X3,y,theta,alpha,iterations)
```

```python
# Plot X1
plt.figure()
plt.scatter(X1[:,[1]], y,  color='black') 

# Plot a red line corrensponding to the model for X1.
plt.plot(X1[:,[1]], np.sum(np.multiply(X1,theta1), axis=1), color='red', linewidth=1)

plt.xlabel("X1")
plt.ylabel("y")
plt.title("Regression Model for X1")

# Plot X2
plt.figure()
plt.scatter(X2[:,[1]], y,  color='black') 

# Plot a red line corrensponding to the model for X2.
plt.plot(X2[:,[1]], np.sum(np.multiply(X2,theta2), axis=1), color='red', linewidth=1)

plt.xlabel("X2")
plt.ylabel("y")
plt.title("Regression Model for X2")

# Plot X3
plt.figure()
plt.scatter(X3[:,[1]], y,  color='black') 

# Plot a red line corrensponding to the model for X3.
plt.plot(X3[:,[1]], np.sum(np.multiply(X3,theta3), axis=1), color='red', linewidth=1)

plt.xlabel("X3")
plt.ylabel("y")
plt.title("Regression Model for X1")

# Plotting the iterations vs the cost history
plt.figure()
plt.plot(cost1_history[0:len(cost1_history)], color='blue', linewidth=1)
plt.xlabel("Iterations")
plt.ylabel("J1")
plt.title("J1 vs. Iterations");

plt.figure()
plt.plot(cost2_history[0:len(cost2_history)], color='blue', linewidth=1)
plt.xlabel("Iterations")
plt.ylabel("J2")
plt.title("J2 vs. Iterations");

plt.figure()
plt.plot(cost3_history[0:len(cost3_history)], color='blue', linewidth=1)
plt.xlabel("Iterations")
plt.ylabel("J3")
plt.title("J3 vs. Iterations");
```
