```python
# import dataset
# Map stringed input to numerical values 
# Found this tutorial helpful https://www.datacamp.com/community/tutorials/categorical-data

import numpy as np 
import matplotlib.pyplot as plt 
import pandas as pd
import seaborn as sns 
import warnings 

import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler 
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from sklearn import metrics
from matplotlib.colors import ListedColormap
from sklearn.decomposition import PCA 
from sklearn.svm import SVC 
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
from sklearn.model_selection import KFold
from imblearn.under_sampling import RandomUnderSampler
from sklearn.naive_bayes import GaussianNB 
```

# Importing and Viewing Data Before Pre-Processing

```python
# Reading in the dataset
bank = pd.DataFrame(pd.read_csv(r'bank-additional/bank-additional-full.csv',sep=';'))
#bank = pd.DataFrame(pd.read_csv(r'/content/drive/MyDrive/Colab Notebooks/bank-additional-full.csv',sep=';'))
# Separating the categorical data
bank_cat = bank.select_dtypes(include=['object']).copy()
```

```python
# Finding the frequency distribution of the some of the categorical data
%matplotlib inline
fig, axs = plt.subplots(1,1,figsize=(7,5))
job_count = bank_cat['job'].value_counts()
sns.set(style="darkgrid")
sns.barplot(job_count.index, job_count.values, alpha=0.9)
plt.title('Frequency Distribution of Jobs')
plt.ylabel('Number of Occurrences', fontsize=12)
plt.xlabel('Jobs', fontsize=12)
fig.autofmt_xdate()
plt.savefig('figures//jobs_hist',dpi=1600)
plt.show()

# %matplotlib inline
fig, axs = plt.subplots(1,1,figsize=(7,5))
marital_count = bank_cat['marital'].value_counts()
sns.set(style="darkgrid")
sns.barplot(marital_count.index, marital_count.values, alpha=0.9)
plt.title('Frequency Distribution of Marital Status')
plt.ylabel('Number of Occurrences', fontsize=12)
plt.xlabel('Marital Status', fontsize=12)
fig.autofmt_xdate()
plt.savefig('figures//marry_hist',dpi=1600)
plt.show()

# %matplotlib inline
fig, axs = plt.subplots(1,1,figsize=(7,6))
education_count = bank_cat['education'].value_counts()
sns.set(style="darkgrid")
sns.barplot(education_count.index, education_count.values, alpha=0.9)
plt.title('Frequency Distribution of Education Level')
plt.ylabel('Number of Occurrences', fontsize=12)
plt.xlabel('Education Level', fontsize=12)
fig.autofmt_xdate()
plt.savefig('figures//edu_hist',dpi=1600)
plt.show()

# %matplotlib inline
fig, axs = plt.subplots(1,1,figsize=(7,5))
y_count = bank_cat['y'].value_counts()
sns.set(style="darkgrid")
sns.barplot(y_count.index, y_count.values, alpha=0.9)
plt.title('Frequency Distribution of the Outcome')
plt.ylabel('Number of Occurrences', fontsize=12)
plt.xlabel('Outcome', fontsize=12)
fig.autofmt_xdate()
plt.savefig('figures//outcome_hist',dpi=1600)
plt.show()
```

# Feature Processing

```python
# Encoding the categories entries to numericl entries
for (category, data) in bank_cat.iteritems():
  bank[category] = bank_cat[category].astype('category').cat.codes

# Splitting the data and scaling the data
X = bank.iloc[:,:20].values 
Y = bank.iloc[:, 20].values 

rus = RandomUnderSampler(random_state=0, replacement=True)# fit predictor and target variable
X, Y = rus.fit_resample(X, Y)

# Scaling the features
sc_X = StandardScaler() 
std_X = sc_X.fit_transform(X)
```

# Logistic Regression Without Feature Extraction and K-Fold Cross Validation

```python
# Training the logistics model first without PCA or LDA or K fold
X_train, X_test, Y_train, Y_test = train_test_split(std_X, Y, test_size = 0.20, random_state=0)

# Creating the logistic model and fitting it
classifier = LogisticRegression(random_state=0) 
classifier.fit(X_train, Y_train)

# Creating predictions with the test data
Y_pred = classifier.predict(X_test)

# Computing the metrics for the model
cnf_matrix = confusion_matrix(Y_test, Y_pred) 

print("Accuracy:",metrics.accuracy_score(Y_test, Y_pred)) 
print("Precision:",metrics.precision_score(Y_test, Y_pred)) 
print("Recall:",metrics.recall_score(Y_test, Y_pred)) 
```

```python
# Confusion matrix
class_names=[0,1] # name  of classes 
fig, ax = plt.subplots() 
tick_marks = np.arange(len(class_names)) 
plt.xticks(tick_marks, class_names) 
plt.yticks(tick_marks, class_names) 
# create heatmap 
sns.heatmap(pd.DataFrame(cnf_matrix), annot=True, cmap="YlGnBu" ,fmt='g') 
ax.xaxis.set_label_position("top") 
plt.tight_layout() 
plt.title('Confusion matrix for Logistic Regression Without Feature Extraction', y=1.1) 
plt.ylabel('Actual label') 
plt.xlabel('Predicted label');
plt.savefig('figures//regression_unchanged',dpi=1600)
plt.show()
```

# Logistic Regression with PCA

```python
# Training the logistics model with PCA
# First determinng the best N
accuracy = []
precision = []
recall = []

k_fold = KFold(n_splits=5, random_state=0, shuffle=True)

for i in range(20):
    pca = PCA(n_components=i+1)
    principalComponents = pca.fit_transform(std_X)
    principalDf = pd.DataFrame(data = principalComponents)
    principalNp = principalDf.to_numpy()
    
    predicted_targets = np.array([])
    actual_targets = np.array([])

    for train_ix, test_ix in k_fold.split(principalNp):
        X_train, Y_train, X_test, Y_test = principalNp[train_ix], Y[train_ix], principalNp[test_ix], Y[test_ix]

        # Fit the classifier
        classifier = LogisticRegression(random_state=0)
        classifier.fit(X_train, Y_train)

        # Predict the labels of the test set samples
        predicted_labels = classifier.predict(X_test)

        predicted_targets = np.append(predicted_targets, predicted_labels)
        actual_targets = np.append(actual_targets, Y_test)

    accuracy.append(metrics.accuracy_score(predicted_targets, actual_targets))
    precision.append(metrics.precision_score(predicted_targets, actual_targets))
    recall.append(metrics.recall_score(predicted_targets, actual_targets))

# Plotting the accuracy, precision, and recall against the iterations
plt.figure()
plt.plot(range(20),accuracy)
plt.xlabel('Iterations')
plt.ylabel('Accuracy')
plt.title('Accuracy vs the Iterations')
plt.xticks(range(20))
plt.savefig('figures//accuracy',dpi=1600)
plt.show()

plt.figure()
plt.plot(range(20),precision)
plt.xlabel('Iterations')
plt.ylabel('Precision')
plt.title('Precision vs the Iterations')
plt.xticks(range(20))
plt.savefig('figures//precision',dpi=1600)
plt.show()

plt.figure()
plt.plot(range(20),recall)
plt.xlabel('Iterations')
plt.ylabel('Recall')
plt.title('Recall vs the Iterations')
plt.xticks(range(20))
plt.savefig('figures//recall',dpi=1600)
plt.show()
```

# Logistics Regression With the LDA and K-Fold

```python
# Logistics regression without pca or lda is better.

lda = LDA(n_components=1)
lda_t = lda.fit_transform(std_X,Y)

predicted_targets = np.array([])
actual_targets = np.array([])

k_fold = KFold(n_splits=5, random_state=0, shuffle=True)

for train_ix, test_ix in k_fold.split(std_X):
    X_train, Y_train, X_test, Y_test = std_X[train_ix], Y[train_ix], std_X[test_ix], Y[test_ix]

    # Fit the classifier
    lda.fit(X_train, Y_train)

    # Predict the labels of the test set samples
    predicted_labels = lda.predict(X_test)

    predicted_targets = np.append(predicted_targets, predicted_labels)
    actual_targets = np.append(actual_targets, Y_test)


# Computing the metrics for the model
cnf_matrix = confusion_matrix(predicted_targets, actual_targets) 

print("Accuracy:",metrics.accuracy_score(predicted_targets, actual_targets)) 
print("Precision:",metrics.precision_score(predicted_targets, actual_targets)) 
print("Recall:",metrics.recall_score(predicted_targets, actual_targets))

# Confusion matrix
class_names=[0,1] # name  of classes 
fig, ax = plt.subplots() 
tick_marks = np.arange(len(class_names)) 
plt.xticks(tick_marks, class_names) 
plt.yticks(tick_marks, class_names) 
# create heatmap 
sns.heatmap(pd.DataFrame(cnf_matrix), annot=True, cmap="YlGnBu" ,fmt='g') 
ax.xaxis.set_label_position("top") 
plt.tight_layout() 
plt.title('Confusion matrix for Logistic Regression with PCA and K-Fold', y=1.1) 
plt.ylabel('Actual label') 
plt.xlabel('Predicted label');
plt.savefig('figures//regression_pca_kfold',dpi=1600)
plt.show()
```

# Naive Bayes with LDA and K-Fold

```python
# Training a Naive Bayes model with LDA
lda = LDA(n_components=1)
lda_t = lda.fit_transform(std_X,Y)

predicted_targets = np.array([])
actual_targets = np.array([])

k_fold = KFold(n_splits=5, random_state=0, shuffle=True)

for train_ix, test_ix in k_fold.split(lda_t):
    train_x, train_y, test_x, test_y = lda_t[train_ix], Y[train_ix], lda_t[test_ix], Y[test_ix]

    # Creating the Naive Bayes model and fitting it
    lda.fit(train_x,train_y)

    # Predict the labels of the test set samples
    predicted_labels = lda.predict(test_x)

    predicted_targets = np.append(predicted_targets, predicted_labels)
    actual_targets = np.append(actual_targets, test_y)


# Computing the metrics for the model
cnf_matrix = confusion_matrix(predicted_targets, actual_targets) 

print("Accuracy:",metrics.accuracy_score(predicted_targets, actual_targets)) 
print("Precision:",metrics.precision_score(predicted_targets, actual_targets)) 
print("Recall:",metrics.recall_score(predicted_targets, actual_targets))

# Confusion matrix
class_names=[0,1] # name  of classes 
fig, ax = plt.subplots() 
tick_marks = np.arange(len(class_names)) 
plt.xticks(tick_marks, class_names) 
plt.yticks(tick_marks, class_names) 
# create heatmap 
sns.heatmap(pd.DataFrame(cnf_matrix), annot=True, cmap="YlGnBu" ,fmt='g') 
ax.xaxis.set_label_position("top") 
plt.tight_layout() 
plt.title('Confusion matrix for Naive Bayes with LDA and K-Fold', y=1.1) 
plt.ylabel('Actual label') 
plt.xlabel('Predicted label');
plt.savefig('figures//naive_bayes_lda',dpi=1600)
plt.show()
```

# Support Vector Machine with linear kernel and K-Fold and LDA

```python
# Reading in the smaller dataset for svm
bank = pd.DataFrame(pd.read_csv(r'bank-additional/bank-additional.csv',sep=';'))
#bank = pd.DataFrame(pd.read_csv(r'/content/drive/MyDrive/Colab Notebooks/bank-additional-full.csv',sep=';'))
# Separating the categorical data
bank_cat = bank.select_dtypes(include=['object']).copy()

# Encoding the categories entries to numericl entries
for (category, data) in bank_cat.iteritems():
  bank[category] = bank_cat[category].astype('category').cat.codes

# Splitting the data and scaling the data
X = bank.iloc[:,:20].values 
Y = bank.iloc[:, 20].values 

rus = RandomUnderSampler(random_state=0, replacement=True)# fit predictor and target variable
X, Y = rus.fit_resample(X, Y)

# Scaling the features
sc_X = StandardScaler() 
std_X = sc_X.fit_transform(X)
```

```python
# Using a linear kernel
lda = LDA(n_components=1)
lda_t = lda.fit_transform(std_X,Y)

predicted_targets = np.array([])
actual_targets = np.array([])

k_fold = KFold(n_splits=5, random_state=0, shuffle=True)

# Creating the Naive Bayes model and fitting it
for train_ix, test_ix in k_fold.split(lda_t):
    X_train, Y_train, X_test, Y_test = lda_t[train_ix], Y[train_ix], lda_t[test_ix], Y[test_ix]
    
    model = SVC(kernel='linear', C=1E10) 
    model.fit(X_train, Y_train)

    # Predict the labels of the test set samples
    predicted_labels = model.predict(X_test)

    predicted_targets = np.append(predicted_targets, predicted_labels)
    actual_targets = np.append(actual_targets, Y_test)

# Creating predictions with the test data
Y_pred = model.predict(X_test)

cnf_matrix = confusion_matrix(actual_targets, predicted_targets) 
print("Accuracy:",metrics.accuracy_score(predicted_targets, actual_targets)) 
print("Precision:",metrics.precision_score(predicted_targets, actual_targets)) 
print("Recall:",metrics.recall_score(predicted_targets, actual_targets))

class_names=[0,1] # name  of classes 
fig, ax = plt.subplots() 
tick_marks = np.arange(len(class_names)) 
plt.xticks(tick_marks, class_names) 
plt.yticks(tick_marks, class_names) 
# create heatmap 
sns.heatmap(pd.DataFrame(cnf_matrix), annot=True, cmap="YlGnBu" ,fmt='g') 
ax.xaxis.set_label_position("top") 
plt.tight_layout() 
plt.title('Confusion matrix for Support Vector Machine with LDA and K-Fold and Liener Kernel', y=1.1) 
plt.ylabel('Actual label') 
plt.xlabel('Predicted label');
plt.savefig('figures//svm-linear',dpi=1600)
plt.show()
```
